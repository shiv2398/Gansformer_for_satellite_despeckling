```python

train_data_images=os.listdir("../input/satellite-despeckling-train-dataset/clean_1c")
test_data_images=os.listdir("../input/satellite-despeckling-train-dataset/noisy_1c")
```

```python
path1="../input/satellite-despeckling-train-dataset/clean_1c/"
path2="../input/satellite-despeckling-train-dataset/noisy_1c/"
```

```python
def data_ex(path,list1):
    i=0
    train_data=[]
    for im in list1:
      img=image.load_img(path+im,target_size=(256,256))
      if i<=400:
          img=image.img_to_array(img)
          img=img/255
          train_data.append(img)
      else :
        pass
      i+=1
    return train_data


  
  

```

```python
train_data=data_ex(path1,train_data_images)
test_data=data_ex(path2,test_data_images)
```

```python
def paths_make(data,path):
    image_paths=[]
    i=1
    for p in data:
     if i<=400:
         image_paths.append(str(path+"/"+p))
     else:
        break
     i+=1
    return image_paths
```

```python
#checking wheather images are greyscale or not 
from PIL import Image, ImageStat

def is_grayscale(path="image.jpg"):
    im = Image.open(path).convert("RGB")
    stat = ImageStat.Stat(im)
    return sum(stat.sum)/3 == stat.sum[0]
```

```python
def check_gray_scale(image_p,title):
    grey_scale_sum=[]
    for p in image_p:
        grey_scale_sum.append(is_grayscale(p))

    plt.hist(grey_scale_sum)
    plt.xlabel("gray_scale_sum")
    plt.ylabel("values_count")
    plt.title(title)
```

```python
tr_paths=paths_make(train_data_images,path1)
te_paths=paths_make(train_data_images,path2)
check_gray_scale(tr_paths,"train_data")
check_gray_scale(te_paths,"test_data")
```

```python
  def intes_scatter(train_data,title,color):
    train_data_mean=[]
    train_data_std=[]
    for p in train_data:
      train_data_mean.append(np.mean(p))
      train_data_std.append(np.std(p))
    plt.scatter(train_data_std,train_data_mean,color=color)
    plt.xlabel("mean")
    plt.ylabel("std_deviation")
    plt.title(title)
```

```python
def histogram(paths,title):
    gray_img = cv2.imread(paths)
    hist_r = cv2.calcHist([gray_img],[0],None,[256],[0,256])
    plt.hist(gray_img.ravel(),256,[0,256])
    plt.title(title+'Histogram')
    plt.show()

```
